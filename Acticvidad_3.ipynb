Problema 3: Recibo de Compra (Anidación y Acceso) Estás diseñando la estructura de datos para un recibo de compra, que debe contener información estructurada y repetitiva.

Tareas Estructura del Recibo: Crea un diccionario llamado recibo con las siguientes claves:

'numero_recibo' (entero).

'fecha' (cadena).

'items' (lista).

El valor de 'items' debe ser una lista de diccionarios, donde cada diccionario representa un ítem con claves como 'nombre', 'cantidad', y 'precio_unitario'.

Llenar con Datos: Agrega al menos dos ítems (diccionarios) a la lista 'items' del recibo.

Cálculo y Acceso:

Accede e imprime el nombre del segundo ítem comprado.

Calcula el subtotal de ese segundo ítem (cantidad * precio_unitario) y almacénalo en una variable.

Modificación Anidada: Incrementa la 'cantidad' del primer ítem en 1.

```python
recibo = {
    'numero_recibo': 1001,
    'fecha': '2024-01-16',
    'items': [
        {'nombre': 'Laptop', 'cantidad': 1, 'precio_unitario': 12000.00},
        {'nombre': 'Mouse', 'cantidad': 2, 'precio_unitario': 250.00},
        {'nombre': 'Teclado', 'cantidad': 1, 'precio_unitario': 350.00}
    ]
}
print("Recibo con ítems:", recibo)
```

Cálculo y Acceso:

Accede e imprime el nombre del segundo ítem comprado.

Calcula el subtotal de ese segundo ítem (cantidad * precio_unitario) y almacénalo en una variable.

Modificación Anidada: Incrementa la 'cantidad' del primer ítem en 1.

```python
# Accede e imprime el nombre del segundo ítem comprado.
segundo_item_nombre = recibo['items'][1]['nombre']
print(f"Nombre del segundo ítem: {segundo_item_nombre}")

# Calcula el subtotal de ese segundo ítem (cantidad * precio_unitario) y almacénalo en una variable.
segundo_item_cantidad = recibo['items'][1]['cantidad']
segundo_item_precio = recibo['items'][1]['precio_unitario']
subtotal_segundo_item = segundo_item_cantidad * segundo_item_precio
print(f"Subtotal del segundo ítem ({segundo_item_nombre}): {subtotal_segundo_item:.2f}")

# Modificación Anidada: Incrementa la 'cantidad' del primer ítem en 1.
recibo['items'][0]['cantidad'] += 1

print("\nRecibo después de la modificación:", recibo)
```
