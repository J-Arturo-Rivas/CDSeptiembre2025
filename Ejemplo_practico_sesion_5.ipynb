```python
# 1. Importar NumPy
import numpy as np

print("--- 1. NumPy: Operaciones Vectoriales ---")

# Datos de ejemplo (precios de productos)
precios_lista = [150, 250, 75, 1200, 30]
iva = 0.16

# Intentar aplicar el IVA a una lista estándar (daría error si no se usa un bucle)
# print(precios_lista * (1 + iva)) # Comentamos para evitar el error de demostración

# Convertir la lista a un Array de NumPy
print(f"Lista original: {precios_lista}")
precios_array = np.array(precios_lista)
print(f"Array original: {precios_array}")

# Operación Vectorial: Multiplicar el array completo por (1 + IVA)
precios_con_iva = precios_array * (1 + iva)

print(f"Precios con IVA aplicado (Operación Vectorial):")
print(precios_con_iva.round(2)) # Usamos .round(2) para mejor visualización
```
