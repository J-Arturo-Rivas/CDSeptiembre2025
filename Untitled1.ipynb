Muestra las primeras 5 filas y la forma (.shape) del DataFrame ventas_df.

```python
import pandas as pd

# Cargar el archivo CSV en un DataFrame
ventas_df = pd.read_csv('ventas.csv')

# Mostrar las primeras 5 filas del DataFrame
print('Primeras 5 filas del DataFrame:')
print(ventas_df.head())

# Mostrar la forma (número de filas y columnas) del DataFrame
print('\nForma del DataFrame (filas, columnas):')
print(ventas_df.shape)
```

Crea un nuevo DataFrame (productos_precios) que solo contenga las columnas Producto y Precio.

```python
productos_precios = ventas_df[['Producto', 'Precio']]
print('Nuevo DataFrame productos_precios:')
print(productos_precios)
```

Define un diccionario (mapa_categorias) que clasifique los productos en categorías ('Hardware').
Ejemplo: 'Laptop': 'Hardware'
print(mapa_categoriaS)

```python
productos_unicos = productos_precios['Producto'].unique()
mapa_categorias = {producto: 'Hardware' for producto in productos_unicos}

# Ajustamos algunas categorías a 'Accesorios' para mayor realismo, siguiendo el ejemplo anterior
# Si se desea que todos sean 'Hardware', se puede omitir este paso.
if 'Cable HDMI' in mapa_categorias: mapa_categorias['Cable HDMI'] = 'Accesorios'
if 'Funda Silicona' in mapa_categorias: mapa_categorias['Funda Silicona'] = 'Accesorios'
if 'Cargador Rápido' in mapa_categorias: mapa_categorias['Cargador Rápido'] = 'Accesorios'
if 'Batería Externa' in mapa_categorias: mapa_categorias['Batería Externa'] = 'Accesorios'
if 'Memoria USB 64GB' in mapa_categorias: mapa_categorias['Memoria USB 64GB'] = 'Accesorios'

print('Diccionario mapa_categorias generado desde productos_precios:')
print(mapa_categorias)
```

Aplicación: Usa el método .map() para aplicar tu diccionario a la columna Producto y crea una nueva columna llamada Categoría en ventas_df.

```python
ventas_df['Categoría'] = ventas_df['Producto'].map(mapa_categorias)

print('DataFrame ventas_df con la nueva columna Categoría:')
print(ventas_df.head())
```

```python
conteo_por_categoria = ventas_df['Categoría'].value_counts()
print('Conteo de productos por categoría:')
print(conteo_por_categoria)
```

Simula un flujo de trabajo que respete la lógica LIFO (Último en Entrar, Primero en Salir):

Setup: Inicializa una lista llamada pila_inventario con ['Revisión A', 'Revisión B'].

Push: Agrega las tareas 'Revisión C' y 'Revisión D' usando la lógica de Pila.

```python
# Setup: Inicializa la pila_inventario
pila_inventario = ['Revisión A', 'Revisión B']
print(f"Pila inicial: {pila_inventario}")

# Push: Agrega tareas usando la lógica de Pila (LIFO)
pila_inventario.append('Revisión C')
pila_inventario.append('Revisión D')
print(f"Pila después de Push: {pila_inventario}")
```

Pop: Ejecuta la última tarea que entró. Guarda el resultado en una variable e imprime qué tarea fue ejecutada

```python
# Pop: Ejecuta la última tarea que entró (LIFO)
tarea_ejecutada = pila_inventario.pop()
print(f"Tarea ejecutada (Pop): {tarea_ejecutada}")
print(f"Pila después de Pop: {pila_inventario}")
```
