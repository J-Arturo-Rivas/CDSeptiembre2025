```python
nombre= input("Ingresa tu nombre")
print(f"Hola {nombre}")
print("Esto es una lista")
lista1=[1,2,3,4,5]
print(lista1)
print("Asi imprimes el primer elemento de la lista")
print(lista1[0])
print("Asi agregas un elemento a la lista")
lista1.append(6)
print(lista1)
print("Esto es una tupla")
tupla1=(1,2,3,4,5)
print("Asi imprimes el primer elemento de la tupla")
print(tupla1[0])
print("Asi imprimes los primeros 3 elementos de una tupla")
print(tupla1[0:3])
print("Esto es un diccionario")
diccionario1={"nombre":"Juan", "edad":20}
print(diccionario1)
print("Asi imprimes el valor de la clave nombre del diccionario")
print(diccionario1["nombre"])
```
