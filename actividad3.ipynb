Problema 3: Recibo de Compra (Anidación y Acceso)
Estás diseñando la estructura de datos para un recibo de compra, que debe contener información estructurada y repetitiva.

Estructura del Recibo: Crea un diccionario llamado recibo con las siguientes claves 'numero_recibo' (entero).

'fecha' (cadena).

'items' (lista).l valor de 'items' debe ser una lista de diccionarios, donde cada diccionario representa un ítem con claves como 'nombre', 'cantidad', y 'precio_unitario'.Llenar con Datos: Agrega al menos dos ítems (diccionarios) a la lista 'items' del recibo.

```python
numero_recibo = 12345
nombre_cliente = 'sael'
recibo = {    'numero': numero_recibo,
    'cliente': nombre_cliente,
    'fecha': '12/02/23',
    'items': [{"nombre":"papel","cantidad":5,"precio":15},{"nombre":"tornillo","cantidad":15,"precio":1}]
}
```



Cálculo y Acceso:

Accede e imprime el nombre del segundo ítem comprado.

Calcula el subtotal de ese segundo ítem (cantidad * precio_unitario) y almacénalo en una variable.

Modificación Anidada: Incrementa la 'cantidad' del primer ítem en 1.

```python
nombre_primer_item = recibo['items'][0]['nombre']
print(f"Nombre del primer ítem: {nombre_primer_item}")

cantidad_primer_item = recibo['items'][0]['cantidad']
precio_primer_item = recibo['items'][0]['precio']
subtotal_primer_item = cantidad_primer_item * precio_primer_item
print(f"Subtotal del primer ítem: {subtotal_primer_item}")
recibo['items'][0]['cantidad'] += 1
print(f"Nueva cantidad del primer ítem: {recibo['items'][0]['cantidad']}")
print("Recibo actualizado:")
print(recibo)
```

```python
# Accede e imprime el nombre del segundo ítem comprado.
nombre_segundo_item = recibo['items'][1]['nombre']
print(f"Nombre del segundo ítem: {nombre_segundo_item}")

# Calcula el subtotal de ese segundo ítem (cantidad * precio_unitario) y almacénalo en una variable.
cantidad_segundo_item = recibo['items'][1]['cantidad']
precio_segundo_item = recibo['items'][1]['precio'] # Assuming 'precio' is the key for unit price
subtotal_segundo_item = cantidad_segundo_item * precio_segundo_item
print(f"Subtotal del segundo ítem: {subtotal_segundo_item}")

# Modificación Anidada: Incrementa la 'cantidad' del primer ítem en 1.
recibo['items'][0]['cantidad'] += 1
print(f"Nueva cantidad del primer ítem: {recibo['items'][0]['cantidad']}")
print("Recibo actualizado:")
print(recibo)
```
