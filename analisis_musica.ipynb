**#üìä Problema 1: Estad√≠sticas B√°sicas y C√°lculos con NumPy**

Calcula las estad√≠sticas fundamentales de las reproducciones musicales utilizando la eficiencia de NumPy.

Pasos:

1.- Conversi√≥n: Extrae la columna reproducciones y convi√©rtela a Array de NumPy

2.- C√°lculos Estad√≠sticos: Usando NumPy, calcula:

  Media, mediana, desviaci√≥n est√°ndar

  Rango intercuart√≠lico (IQR)

Resultados: Muestra todos los c√°lculos formateados

Problema 1: Estad√≠sticas con NumPy

Ejemplo de estructura:

    reproducciones_np = ... # Convertir a NumPy array (PRACTICA ANTERIOR) #5
    media = np.mean(...)
    mediana = np.median(...)

```python
#An√°lisis estad√≠stico de reproducciones
print("AN√ÅLISIS ESTAD√çSTICO DE REPRODUCCIONES.\n")

#Conversi√≥n de columna de reproducciones en un array de NumPy
reproducciones_np = df_music['reproducciones'].to_numpy()

#C√°lculos estad√≠sticos

media_reproducciones = np.mean(reproducciones_np)
mediana_reproducciones = np.median(reproducciones_np)
desviacion_estandar_reproducciones = np.std(reproducciones_np)
max_reproducciones = np.max(reproducciones_np)
min_reproducciones = np.min(reproducciones_np)

#Rango intercuart√≠lico
percentil_25 = np.percentile(reproducciones_np, 25)
percentil_75 = np.percentile(reproducciones_np, 75)
iqr_reproducciones = (percentil_75 - percentil_25)


#Muestra de los c√°lculos formateados.

print("Estad√≠sticas de Reproducciones\n")
print(f"Reproducciones promedio: {media_reproducciones:.2f} ")
print(f"Mediana de Reproducciones: {mediana_reproducciones:.2f}")
print(f"Desviaci√≥n est√°ndar de Reproducciones: {desviacion_estandar_reproducciones:.2f}")
print(f"M√°ximo de Reproducciones: {max_reproducciones:.2f}")
print(f"M√≠nimo de Reproducciones: {min_reproducciones:.2f}")
print("\nInformaci√≥n del rango intercuart√≠lico: \n")
print(f"Percentil 25: {percentil_25:.2f}")
print(f"Percentil 75: {percentil_75:.2f}")
print(f"Rango Intercuart√≠lico (IQR): {iqr_reproducciones:.2f}")
```

```python
# An√°lisis de distribuci√≥n por g√©nero
print('An√°lisis por g√©nero musical\n')

# Agrupar por g√©nero y suma total de reproducciones.

reproducciones_por_genero = df_music.groupby('genero')['reproducciones'].sum()

print(f"Resumen por g√©nero: \n{reproducciones_por_genero}\n")

# Ordenar valores de mayor a menor
reproducciones_por_genero = reproducciones_por_genero.sort_values(ascending=False)

# Visualizaci√≥n - Gr√°fica de barras horizontal
plt.barh(reproducciones_por_genero.index, reproducciones_por_genero.values)
# Configurar t√≠tulo
plt.title('Distribuci√≥n de Reproducciones por G√©nero')
#Configuraci√≥n de etiquetas
plt.xlabel('Reproducciones')
plt.ylabel('G√©nero')
# Mostrar gr√°fica
plt.show()

#Visualizaci√≥n - Gr√°fica de pasteles porcentual

colores = plt.cm.Set3(np.linspace(0, 1, len(reproducciones_por_genero)))
''' Crear gr√°fico de pastel, porcentajes formateados a un decimal,
    inicio de la primera porci√≥n en la parte superior,
    con una separaci√≥n entre fragmentos y sombra.
'''
plt.pie(reproducciones_por_genero.values,
        labels=reproducciones_por_genero.index,
        autopct='%1.1f%%',
        startangle=90,
        colors=colores,
        explode=[0.05] * len(reproducciones_por_genero),

        )
plt.title('Distribuci√≥n de Reproducciones por G√©nero\n')
plt.legend(reproducciones_por_genero.index,
           title="Reproducci√≥n por g√©nero",
           loc="center left",
           )
plt.axis('equal')

plt.show()
```

üîç Problema 3: An√°lisis Temporal y Correlaciones
Investiga la evoluci√≥n temporal y la relaci√≥n entre duraci√≥n y popularidad.

Pasos:

1.-An√°lisis Temporal: Convierte fecha y analiza evoluci√≥n mensual

2.-Gr√°fico de L√≠neas: Tendencia temporal de reproducciones

3.-Gr√°fico de Dispersi√≥n: Relaci√≥n duraci√≥n vs reproducciones

4.-Interpretaci√≥n: Analiza correlaci√≥n entre variables

Problema 3: An√°lisis Temporal y Correlaciones

Ejemplo de estructura:

Convertir columna fecha a datetime

Agrupar por mes y calcular totales

Crear gr√°fico de l√≠neas

Crear gr√°fico de dispersi√≥n

Calcular y mostrar correlaci√≥n

```python
df_music['fecha'] = pd.to_datetime(df_music['fecha'])
df_music['mes'] = df_music['fecha'].dt.month
reproducciones_por_mes = df_music.groupby('mes')['reproducciones'].sum()
plt.plot(reproducciones_por_mes.index, reproducciones_por_mes.values, marker='o')
plt.xlabel('Mes')
plt.ylabel('Reproducciones')
plt.title('Tendencia Temporal de Reproducciones')
plt.grid(True)
plt.show()

plt.scatter(df_music['duracion_min'], df_music['reproducciones'])
plt.xlabel('Duraci√≥n')
plt.ylabel('Reproducciones')
plt.title('Relaci√≥n entre Duraci√≥n y Reproducciones')
plt.grid(True)
plt.show()

correlacion = df_music['duracion_min'].corr(df_music['reproducciones'])
print(f"Correlaci√≥n entre Duraci√≥n y Reproducciones: {correlacion:.2f}")

print('Con base en el resultado de la correlaci√≥n entre duraci√≥n y reproducciones se asume que no existe una correlaci√≥n, \n ya que como se observa en el Gr√°fico 2 "Relaci√≥n entre Duraci√≥n y Reproducciones, \n muestra una dispersi√≥n no perdecible ni vinculativa.')
```

üéµ Problema 4: Top Canciones y Artistas
Identifica las canciones y artistas m√°s populares usando operaciones de Pandas.

Pasos:

Top Canciones: Encuentra las 10 canciones con m√°s reproducciones

Top Artistas: Identifica los 5 artistas m√°s escuchados

Visualizaci√≥n Combinada: Crea visualizaci√≥n con ambos rankings

Problema 4: Top Canciones y Artistas
Ejemplo de estructura:

```python
top_canciones = df_music.groupby('cancion')['reproducciones'].sum().sort_values(ascending=False).head(10)
top_artistas = df_music.groupby('artista')['reproducciones'].sum().sort_values(ascending=False).head()

print("Top 10 canciones m√°s reproducidas:")
print(top_canciones)
print("\nTop 5 artistas m√°s escuchados:")
print(top_artistas)

plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.barh(top_canciones.index, top_canciones.values )
plt.xlabel('Reproducciones')
plt.ylabel('Canci√≥n')
plt.title('Top_canciones')
plt.grid(False)
plt.show()

plt.subplot(1, 2, 2)
plt.barh(top_artistas.index, top_artistas.values)
plt.xlabel('Reproducciones')
plt.ylabel('Artista')
plt.title('Top_artistas')
plt.grid(False)
plt.show()

```
